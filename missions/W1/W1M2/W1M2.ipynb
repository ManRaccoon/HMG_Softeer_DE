```python
import sqlite3
```

### 데이터베이스 연결

```python
conn = sqlite3.connect('w3schools.db')
cur = conn.cursor()
```

### 앞으로의 쿼리를 쉽게 작성 및 확인하기 위한 query함수 작성

```python
def query(cur, query):
    cur.execute(query)
    rows = cur.fetchall()
    for row in rows:
        print(row)
```

### SQL SELECT : 특정 열만 반환하도록 함

```python
query(cur, 'SELECT CustomerName, City FROM Customers;')
```

### Customer의 모든 열 반환

```python
query(cur, 'SELECT * FROM Customers;')
```

### SQL SELECT DISTINCT : 중복 요소를 제거한 쿼리 결과 반환

```python
print('DISTINCT 포함 쿼리 개수')
query(cur, 'SELECT COUNT(DISTINCT Country) FROM Customers;')
print('DISTINCT 미포함 쿼리 개수')
query(cur, 'SELECT COUNT(Country) FROM Customers;')
print('DISTINCT 결과 나라명 출력')
query(cur, 'SELECT DISTINCT Country FROM Customers;')
```

### SQL WHERE : 조건 설정

```python
# Mexico의 Customer의 모든 데이터를 전부 가져오시오
query(cur, 'SELECT * FROM Customers WHERE Country="Mexico";')
```

```python
# 고객ID가 1인 고객의 모든 정보를 가져오시오(text는 아포스트로피(') 사용, Numeric은 사용하지 않음)
query(cur, 'SELECT * FROM Customers WHERE CustomerID=1;')
```

```python
# 고객 ID가 80 초과인 고객의 모든 정보를 가져오시오(WHERE문에는 연산자도 설정 가능하다)
query(cur, 'SELECT * FROM Customers WHERE CustomerID > 80;')
```

### SQL ORDER BY : 결과값을 오름차순, 내림차순으로 정렬

```python
# 가격 순으로 정렬하여 모든 제품의 정보를 가져오시오(아무것도 적지 않았을 때는 오름차순)
query(cur, 'SELECT * FROM Products ORDER BY Price;')
```

```python
# 가격을 내림차순으로 정렬하여 모든 제품의 정보를 가져오시오(DESC는 내림차순, ASC는 오름차순)
query(cur, 'SELECT * FROM Products ORDER BY Price DESC;')
```

```python
# 제품 이름을 알파벳순으로 정렬하여 모든 제품의 정보를 가져오시오(데이터가 text형식이면 알파벳 순으로 정렬함)
query(cur, 'SELECT * FROM Products ORDER BY ProductName;')
```

```python
# 제품 이름을 알파벳 내림차순으로 정렬하여 모든 제품의 정보를 가져오시오(text 형식 또한 DESC를 명시하면 내림차순으로 변경, 오름차순은 ASC)
query(cur, 'SELECT * FROM Products ORDER BY ProductName DESC;')
```

```python
# 고객의 모든 정보를 나라, 이름으로 정렬하여 가져오시오(두 열을 함께 정렬 가능, 앞에 있는 부분 먼저 정렬 이후 두번째 열로 정렬된다.)
query(cur, 'SELECT * FROM Customers ORDER BY Country, CustomerName;')
```

```python
# 고객의 모든 정보를 나라는 오름차순, 이름은 내림차순으로 정렬하여 가져오시오(두 열의 정렬 방향을 다르게 설정할 수 있다. 앞의 열 요소가 먼저 정렬되는건 똑같습니다.)
query(cur, 'SELECT * FROM Customers ORDER BY Country ASC, CustomerName DESC;')
```

### SQL AND : WHERE문 내에서 두 조건이 모두 만족하는 것을 요구할 때

```python
# 스페인에 살고 G로 시작하는 고객의 모든 정보를 가져오시오(여러 조건을 하나의 WHERE문에 요구할 때 AND 사용.)
query(cur, 'SELECT * FROM Customers WHERE Country = "Spain" AND CustomerName LIKE "G%";')
```

```python
# 스페인 리우에 살고 고객ID가 50 초과인 고객의 모든 정보를 가져오시오(여러 조건을 하나의 WHERE문에 요구할 때 AND 사용.)
query(cur, 'SELECT * FROM Customers WHERE Country = "Brazil" AND City = "Rio de Janeiro" AND CustomerID > 50;')
```

```python
# 스페인에 살고 고객 이름이 G나 R로 시작하는 고객의 모든 정보를 가져오시오(AND, OR등을 이용한 혼합조건을 통하여 조건에 맞는 결과 반환)
query(cur, 'SELECT * FROM Customers WHERE Country = "Spain" AND (CustomerName LIKE "G%" OR CustomerName LIKE "R%");')
```

```python
# 스페인에 살고 고객 이름이 G나 R로 시작하는 고객의 모든 정보를 가져오시오(위에 적은 코드에서 괄호가 없을 때 어떤 식으로 결과가 달라지는지 확인 -> G로 시작하는 스페인 사람과, R로 시작하는 모든 국가 사람의 결과가 출력되게 된다.)
query(cur, 'SELECT * FROM Customers WHERE Country = "Spain" AND CustomerName LIKE "G%" OR CustomerName LIKE "R%";')
```

### SQL OR : 조건 중 하나라도 만족하는 경우를 요구할 때

```python
# 독일에 살거나 스페인에 사는 고객의 모든 정보를 가져오시오(둘 중 하나라도 조건에 맞는 결과를 가져온다.)
query(cur, 'SELECT * FROM Customers WHERE Country = "Germany" OR Country = "Spain";')
```

```python
# 베를린에 살거나 G로 시작하는 이름을 가지거나 노르웨이에 사는 고객의 모든 정보를 가져오시오
query(cur, 'SELECT * FROM Customers WHERE City = "Berlin" OR CustomerName LIKE "G%" OR Country = "Norway";')
```

```python
# 스페인에 살고 고객 이름이 G나 R로 시작하는 고객의 모든 정보를 가져오시오(AND에 존재하는 Example과 동일, AND, OR등을 이용한 혼합조건을 통하여 조건에 맞는 결과 반환)
query(cur, 'SELECT * FROM Customers WHERE Country = "Spain" AND (CustomerName LIKE "G%" OR CustomerName LIKE "R%");')
```

```python
# 스페인에 살고 고객 이름이 G나 R로 시작하는 고객의 모든 정보를 가져오시오(AND에 존재하는 Example과 동일, 위에 적은 코드에서 괄호가 없을 때 어떤 식으로 결과가 달라지는지 확인 -> G로 시작하는 스페인 사람과, R로 시작하는 모든 국가 사람의 결과가 출력되게 된다.)
query(cur, 'SELECT * FROM Customers WHERE Country = "Spain" AND CustomerName LIKE "G%" OR CustomerName LIKE "R%";')
```

### SQL NOT : 해당 조건을 제외한 모든 결과를 출력

```python
# 스페인에 살지 않는 고객의 모든 정보를 가져오시오(NOT이 한번만 사용될 경우, 전체-조건의 모든 결과라고 생각하면 된다.)
query(cur, 'SELECT * FROM Customers WHERE NOT Country = "Spain";')
```

```python
# 이름이 A로 시작하지 않는 고객의 모든 정보를 가져오시오(NOT은 문자열 검색에서도 같은 방식으로 적용)
query(cur, 'SELECT * FROM Customers WHERE CustomerName NOT LIKE "A%";')
```

```python
# 고객ID가 10~60인 고객을 제외한 고객의 모든 정보를 가져오시오(뒤의 조건을 제외한 나머지가 결과로 나온다는 것을 알 수 있다.)
query(cur, 'SELECT * FROM Customers WHERE CustomerID NOT BETWEEN 10 AND 60;')
```

```python
# 파리와 런던에 사는 고객을 제외한 고객의 모든 정보를 가져오시오(IN을 통해서 여러 조건을 하나의 값과 비교할 수 있는데 그 결과를 NOT을 통해 제외할 수 있다.)
query(cur, 'SELECT * FROM Customers WHERE City NOT IN ("Paris", "London");')
```

```python
# 고객ID가 50을 초과하는 고객을 제외한 고객의 모든 정보를 가져오시오(조건 연산자를 이용한 결과에도 NOT을 적용할 수 있다.)
query(cur, 'SELECT * FROM Customers WHERE NOT CustomerID > 50;')
```

```python
# 고객ID가 50 미만인 고객을 제외한 고객의 모든 정보를 가져오시오(조건 연산자를 이용한 결과에도 NOT을 적용할 수 있다.)
query(cur, 'SELECT * FROM Customers WHERE NOT CustomerID < 50;')
```

### SQL NULL : DB내 값이 비어있는 경우를 나타냄

```python
# 주소가 없는 고객 데이터베이스에서 고객 이름, 수신자 이름, 주소를 출력(현재 데이터베이스에서는 NULL인 데이터가 없기 때문에 출력이 없다.)
instruction = '''
SELECT CustomerName, ContactName, Address
FROM Customers
WHERE Address IS NULL;
'''
query(cur, instruction)
```

```python
# 주소가 없는 고객 데이터베이스에서 고객 이름, 수신자 이름, 주소를 출력(현재 데이터베이스에서는 NULL인 데이터가 없기 때문에 모든 데이터가 조건에 맞게 출력된다.)
instruction = '''
SELECT CustomerName, ContactName, Address
FROM Customers
WHERE Address IS NOT NULL;
'''
query(cur, instruction)
```

### SQL UPDATE : 조건에 맞는 데이터의 요소를 변경

```python
# DB자체의 영구적인 변형이 일어나므로 새롭게 DB를 가져와 환경을 동일하게 만들어 줌
conn = sqlite3.connect('w3schools.db')
cur = conn.cursor()
```

```python
# 고객 데이터베이스에서 고객ID가 1인 고객의 수신자명, 도시를 변경
instruction = '''
UPDATE Customers
SET ContactName = 'Alfred Schmidt', City= 'Frankfurt'
WHERE CustomerID = 1;
'''
query(cur, instruction)
# Update문 이후 DB 확인용
query(cur, 'SELECT * FROM Customers;')
```

### SQL DELETE : 조건에 따른 요소, 테이블 삭제

```python
# 고객 데이터베이스에서 Alfreds Futterkiste라는 고객명을 가진 데이터를 삭제 (조건에 맞는 데이터를 삭제한다)
instruction = '''
DELETE FROM Customers WHERE CustomerName='Alfreds Futterkiste';
'''
query(cur, instruction)
# Delete문 이후 DB 확인용
query(cur, 'SELECT * FROM Customers;')
```

```python
conn.close()
```

```python
# DB자체의 영구적인 변형이 일어나므로 새롭게 DB를 가져와 환경을 동일하게 만들어 줌
conn = sqlite3.connect('w3schools.db')
cur = conn.cursor()
```

```python
# 고객 데이터베이스의 모든 데이터를 삭제 (조건을 적지 않을 경우 테이블의 모든 데이터를 삭제한다)
instruction = '''
DELETE FROM Customers;
'''
query(cur, instruction)
# Delete문 이후 DB 확인용
query(cur, 'SELECT * FROM Customers;')
```

```python
conn.close()
```

```python
# DB자체의 영구적인 변형이 일어나므로 새롭게 DB를 가져와 환경을 동일하게 만들어 줌
conn = sqlite3.connect('w3schools.db')
cur = conn.cursor()
```

```python
# 고객 데이터베이스를 삭제 (DROP문 사용 시 테이블 자체가 삭제된다)
# 위의 instruction은 데이터 전체를 삭제하지만 테이블 자체는 남아있어 출력은 되지 않지만 오류는 발생하지 않고, DROP문은 테이블 자체를 삭제하였기 때문에 오류가 발생한다.
instruction = '''
DROP TABLE Customers;
'''
query(cur, instruction)
# Delete문 이후 DB 확인용
query(cur, 'SELECT * FROM Customers;')
```
